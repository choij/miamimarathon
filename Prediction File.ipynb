```python
import pandas as pd
import numpy as np
import time

Y1NB = pd.read_csv('final_predictions/Y1_NB.csv')
Y2   = pd.read_csv('final_predictions/Y2.csv')
Y1LG = pd.read_csv('final_predictions/Y1_LG.csv')
```

```python
Y1LG.drop([c for c in ['Age','Sex','AvgTime','NumRaces','YrsSinceLast']],inplace=True,axis=1)
```

```python
Y = Y1LG.copy().set_index('Id')
```

```python
Y = Y.join(Y1NB.set_index('ID'))
Y['Y22']=-1.
Y['Y2'] = ""
```

```python
Y2 = Y2.set_index('ID')
```

```python
y2_indices = Y2.index.tolist()
y_indices  = Y.index.tolist()
```

```python
for i in y2_indices:
    if i in y_indices:
        Y.set_value(i,'Y22',Y2.loc[i]['Y2'])
```

```python
all_valid_y2_predictions = Y[Y['Y22']>0]['Y22'].values.tolist()
y2_avg = sum(all_valid_y2_predictions) / len(all_valid_y2_predictions)
```

```python
invalid_y2 = Y[Y['Y22']==-1].index.tolist()
```

```python
for i in invalid_y2:
    Y.set_value(i,'Y22',y2_avg)
```

```python
for i in y_indices:
    t_n = Y.loc[i]['Y22']
    t   = np.round(t_n* (30282.-7942.)+7942.)
    t_s = time.strftime('%H:%M:%S', time.gmtime(t))
    Y.set_value(i,'Y22',t)
    Y.set_value(i,'Y2',t_s)
```

```python
Y.drop('Y22',inplace=True,axis=1)
```
